## 3. Standardize columns and infer semester / year / round
This makes the uploaded data easier to use for dashboard visuals.

```python
def clean_col_name(col):
    col = str(col).strip()
    col = re.sub(r"\s+", " ", col)
    return col

def infer_term_info(filename):
    info = {
        "Semester": "Unknown",
        "Year": "Unknown",
        "Round": "Uploaded Round"
    }

    if not filename:
        return info

    name = Path(filename).name

    short_match = re.search(r'\b([SF])\s*[-_ ]?(\d{2})\b', name, flags=re.I)
    if short_match:
        term_code, yy = short_match.groups()
        info["Semester"] = "Spring" if term_code.upper() == "S" else "Fall"
        info["Year"] = f"20{yy}"

    long_match = re.search(r'\b(Spring|Fall)\s*[-_ ]?(20\d{2}|\d{2})\b', name, flags=re.I)
    if long_match:
        semester, year = long_match.groups()
        info["Semester"] = semester.title()
        info["Year"] = year if len(year) == 4 else f"20{year}"

    round_match = re.search(r'\bRR\s*[-_ ]?(\d+)\b', name, flags=re.I)
    if not round_match:
        round_match = re.search(r'\bRound\s*[-_ ]?(\d+)\b', name, flags=re.I)
    if round_match:
        info["Round"] = f"Round {round_match.group(1)}"

    return info

df = df_input.copy()
df.columns = [clean_col_name(c) for c in df.columns]
inferred = infer_term_info(uploaded_filename)

print("Inferred context from filename:")
print(inferred)
```

## 4. Normalize the funding fields
This cell ensures the dashboard always has the columns it needs.

```python
required_base_columns = {
    "CIO Name": "Unknown CIO",
    "CIO Category": "Uncategorized",
    "Email": np.nan,
    "Comments": np.nan,
    "Amount Requested": 0.0,
    "Amount Approved": 0.0,
    "Amount Allocated": 0.0,
    "Approval Rate": np.nan,
    "Allocation Rate": np.nan
}

for col, default_value in required_base_columns.items():
    if col not in df.columns:
        df[col] = default_value

possible_round_cols = ["Round", "Funding Round", "Round Number"]
possible_semester_cols = ["Semester", "Term"]
possible_year_cols = ["Year", "Funding Year", "Academic Year"]

round_col = next((c for c in possible_round_cols if c in df.columns), None)
semester_col = next((c for c in possible_semester_cols if c in df.columns), None)
year_col = next((c for c in possible_year_cols if c in df.columns), None)

if round_col is None:
    df["Round"] = inferred["Round"]
else:
    df["Round"] = df[round_col].fillna(inferred["Round"]).astype(str)

if semester_col is None:
    df["Semester"] = inferred["Semester"]
else:
    df["Semester"] = df[semester_col].fillna(inferred["Semester"]).astype(str)

if year_col is None:
    df["Year"] = inferred["Year"]
else:
    df["Year"] = df[year_col].fillna(inferred["Year"]).astype(str)

def to_number(series):
    return (
        series.astype(str)
              .str.replace(r"[^\d.\-]", "", regex=True)
              .replace({"": np.nan, "nan": np.nan, "None": np.nan})
              .astype(float)
    )

money_cols = ["Amount Requested", "Amount Approved", "Amount Allocated"]
for col in money_cols:
    df[col] = to_number(df[col]).fillna(0.0)

for col in ["CIO Name", "CIO Category", "Comments", "Email", "Round", "Semester", "Year"]:
    df[col] = df[col].astype(str).str.strip()

df = df[df["CIO Name"].str.lower().ne("total")]
df = df[df["CIO Name"].str.lower().ne("nan")]
df = df[df["CIO Name"].str.strip().ne("")]

df["Approval Rate"] = np.where(
    df["Amount Requested"] > 0,
    df["Amount Approved"] / df["Amount Requested"],
    np.nan
)

df["Allocation Rate"] = np.where(
    df["Amount Approved"] > 0,
    df["Amount Allocated"] / df["Amount Approved"],
    np.nan
)

df["Allocated Share of Requested"] = np.where(
    df["Amount Requested"] > 0,
    df["Amount Allocated"] / df["Amount Requested"],
    np.nan
)

print(f"Usable rows after normalization: {len(df):,}")
display(df.head())
```

## 5. Add derived fields for visuals and audit checks
These fields power the funnel, breakdowns, and data quality section.

```python
df["Funding Cut"] = (df["Amount Requested"] - df["Amount Allocated"]).clip(lower=0)
df["Approval Gap"] = (df["Amount Requested"] - df["Amount Approved"]).clip(lower=0)

df["Missing Requested"] = df["Amount Requested"].eq(0)
df["Missing Allocated"] = df["Amount Allocated"].eq(0)

df["Missing Contact Info"] = (
    df["Email"].isna() |
    df["Email"].eq("") |
    df["Email"].str.lower().eq("nan")
)

df["Reduced Without Comment"] = (
    (df["Amount Allocated"] < df["Amount Requested"]) &
    (
        df["Comments"].isna() |
        df["Comments"].eq("") |
        df["Comments"].str.lower().eq("nan")
    )
)

df["Audit Flag"] = (
    df["Missing Contact Info"] |
    df["Reduced Without Comment"]
)

for rate_col in ["Approval Rate", "Allocation Rate", "Allocated Share of Requested"]:
    df[rate_col] = df[rate_col].replace([np.inf, -np.inf], np.nan)

print("Derived feature columns added.")
display(df[[
    "CIO Name", "CIO Category", "Round", "Semester", "Year",
    "Amount Requested", "Amount Approved", "Amount Allocated",
    "Funding Cut", "Approval Rate", "Allocation Rate", "Audit Flag"
]].head())
```

```python
summary = pd.DataFrame({
    "Metric": [
        "Applications / Rows",
        "Unique CIOs",
        "Total Requested",
        "Total Approved",
        "Total Allocated",
        "Average Approval Rate",
        "Average Allocation Rate",
        "Rows With Audit Flags"
    ],
    "Value": [
        len(df),
        df["CIO Name"].nunique(),
        df["Amount Requested"].sum(),
        df["Amount Approved"].sum(),
        df["Amount Allocated"].sum(),
        df["Approval Rate"].mean(),
        df["Allocation Rate"].mean(),
        int(df["Audit Flag"].sum())
    ]
})

display(summary)
```

## 7. Create interactive filter controls
These widgets act like the frontend controls for the dashboard.

```python
def dropdown_options(series):
    vals = sorted([x for x in series.dropna().astype(str).unique() if x.strip() and x.lower() != "nan"])
    return ["All"] + vals

category_widget = widgets.Dropdown(
    options=dropdown_options(df["CIO Category"]),
    value="All",
    description="Category:",
    layout=widgets.Layout(width="320px")
)

round_widget = widgets.Dropdown(
    options=dropdown_options(df["Round"]),
    value="All",
    description="Round:",
    layout=widgets.Layout(width="240px")
)

semester_widget = widgets.Dropdown(
    options=dropdown_options(df["Semester"]),
    value="All",
    description="Semester:",
    layout=widgets.Layout(width="240px")
)

year_widget = widgets.Dropdown(
    options=dropdown_options(df["Year"]),
    value="All",
    description="Year:",
    layout=widgets.Layout(width="220px")
)

search_widget = widgets.Text(
    value="",
    placeholder="Type a CIO name",
    description="CIO Search:",
    layout=widgets.Layout(width="420px")
)

top_n_widget = widgets.IntSlider(
    value=10,
    min=5,
    max=20,
    step=1,
    description="Top N:",
    layout=widgets.Layout(width="320px")
)

refresh_button = widgets.Button(
    description="Refresh Dashboard",
    button_style="success",
    icon="refresh"
)

reset_button = widgets.Button(
    description="Reset Filters",
    button_style="warning",
    icon="undo"
)

filters_box = widgets.VBox([
    widgets.HTML("<h3 style='margin: 0 0 6px 0;'>Dashboard Filters</h3>"),
    widgets.HBox([category_widget, round_widget]),
    widgets.HBox([semester_widget, year_widget]),
    widgets.HBox([search_widget, top_n_widget]),
    widgets.HBox([refresh_button, reset_button])
])

display(filters_box)
```

## 8. Build helper functions for the dashboard
This is where the notebook turns into a frontend-style experience.

```python
def format_currency(x):
    return f"${x:,.0f}"

def format_percent(x):
    return "N/A" if pd.isna(x) else f"{x:.1%}"

def apply_filters(data):
    filtered = data.copy()

    if category_widget.value != "All":
        filtered = filtered[filtered["CIO Category"] == category_widget.value]

    if round_widget.value != "All":
        filtered = filtered[filtered["Round"] == round_widget.value]

    if semester_widget.value != "All":
        filtered = filtered[filtered["Semester"] == semester_widget.value]

    if year_widget.value != "All":
        filtered = filtered[filtered["Year"] == year_widget.value]

    search_value = search_widget.value.strip().lower()
    if search_value:
        filtered = filtered[filtered["CIO Name"].str.lower().str.contains(search_value, na=False)]

    return filtered

def make_kpi_html(filtered):
    total_requested = filtered["Amount Requested"].sum()
    total_approved = filtered["Amount Approved"].sum()
    total_allocated = filtered["Amount Allocated"].sum()
    unique_cios = filtered["CIO Name"].nunique()
    app_count = len(filtered)
    avg_approval = filtered["Approval Rate"].mean()
    avg_allocation = filtered["Allocation Rate"].mean()

    html = f"""
    <div style='display:flex; flex-wrap:wrap; gap:18px; margin:8px 0 18px 0;'>
        <div style='background:#4F81BD; color:white; padding:20px; border-radius:14px; min-width:250px; box-shadow:0 4px 10px rgba(0,0,0,0.12);'>
            <div style='font-size:18px;'>Total Requested</div>
            <div style='font-size:42px; font-weight:700;'>{format_currency(total_requested)}</div>
        </div>
        <div style='background:#5AA469; color:white; padding:20px; border-radius:14px; min-width:250px; box-shadow:0 4px 10px rgba(0,0,0,0.12);'>
            <div style='font-size:18px;'>Total Approved</div>
            <div style='font-size:42px; font-weight:700;'>{format_currency(total_approved)}</div>
        </div>
        <div style='background:#D9A441; color:black; padding:20px; border-radius:14px; min-width:250px; box-shadow:0 4px 10px rgba(0,0,0,0.12);'>
            <div style='font-size:18px;'>Total Allocated</div>
            <div style='font-size:42px; font-weight:700;'>{format_currency(total_allocated)}</div>
        </div>
        <div style='background:#F3F4F6; color:#111827; padding:20px; border-radius:14px; min-width:250px; box-shadow:0 4px 10px rgba(0,0,0,0.08);'>
            <div style='font-size:18px;'>Applications / CIOs</div>
            <div style='font-size:20px; font-weight:600;'>{app_count:,} applications</div>
            <div style='font-size:20px; font-weight:600;'>{unique_cios:,} CIOs</div>
            <div style='margin-top:10px; color:#4B5563;'>Avg approval: {format_percent(avg_approval)}</div>
            <div style='color:#4B5563;'>Avg allocation: {format_percent(avg_allocation)}</div>
        </div>
    </div>
    """
    return html

def make_round_chart(filtered):
    group = (
        filtered.groupby("Round", dropna=False)[["Amount Requested", "Amount Approved", "Amount Allocated"]]
        .sum()
        .reset_index()
    )

    if group.empty:
        return None

    long_group = group.melt(
        id_vars="Round",
        value_vars=["Amount Requested", "Amount Approved", "Amount Allocated"],
        var_name="Metric",
        value_name="Amount"
    )

    fig = px.bar(
        long_group,
        x="Round",
        y="Amount",
        color="Metric",
        barmode="group",
        title="Requested vs Approved vs Allocated by Round"
    )
    fig.update_layout(height=460, xaxis_title="Round", yaxis_title="Amount ($)")
    return fig

def make_funnel_chart(filtered):
    totals = pd.DataFrame({
        "Stage": ["Requested", "Approved", "Allocated"],
        "Amount": [
            filtered["Amount Requested"].sum(),
            filtered["Amount Approved"].sum(),
            filtered["Amount Allocated"].sum()
        ]
    })

    fig = go.Figure(go.Funnel(
        y=totals["Stage"],
        x=totals["Amount"],
        textinfo="value+percent initial"
    ))
    fig.update_layout(title="Funding Funnel", height=420)
    return fig

def make_category_chart(filtered):
    group = (
        filtered.groupby("CIO Category", dropna=False)
        .agg(
            Requested=("Amount Requested", "sum"),
            Approved=("Amount Approved", "sum"),
            Allocated=("Amount Allocated", "sum"),
            Avg_Approval_Rate=("Approval Rate", "mean")
        )
        .reset_index()
        .sort_values("Requested", ascending=False)
    )

    if group.empty:
        return None

    fig = px.bar(
        group,
        x="CIO Category",
        y="Requested",
        hover_data=["Approved", "Allocated", "Avg_Approval_Rate"],
        title="Requested Funding by CIO Category"
    )
    fig.update_layout(height=460, xaxis_title="Category", yaxis_title="Requested Amount ($)")
    return fig

def make_top_cios_chart(filtered):
    top_n = top_n_widget.value
    group = (
        filtered.groupby("CIO Name", dropna=False)
        .agg(
            Requested=("Amount Requested", "sum"),
            Approved=("Amount Approved", "sum"),
            Allocated=("Amount Allocated", "sum"),
            Funding_Cut=("Funding Cut", "sum")
        )
        .reset_index()
        .sort_values("Funding_Cut", ascending=False)
        .head(top_n)
    )

    if group.empty:
        return None

    fig = px.bar(
        group,
        x="Funding_Cut",
        y="CIO Name",
        orientation="h",
        title=f"Top {top_n} CIOs by Funding Cut",
        hover_data=["Requested", "Approved", "Allocated"]
    )
    fig.update_layout(height=520, xaxis_title="Funding Cut ($)", yaxis_title="CIO")
    fig.update_yaxes(categoryorder="total ascending")
    return fig

def make_audit_summary(filtered):
    summary = pd.DataFrame({
        "Audit Metric": [
            "Rows with missing contact info",
            "Rows reduced without comment",
            "Rows flagged overall"
        ],
        "Count": [
            int(filtered["Missing Contact Info"].sum()),
            int(filtered["Reduced Without Comment"].sum()),
            int(filtered["Audit Flag"].sum())
        ]
    })
    return summary

def make_flagged_table(filtered):
    flagged = filtered.loc[filtered["Audit Flag"], [
        "CIO Name", "CIO Category", "Round", "Semester", "Year",
        "Amount Requested", "Amount Approved", "Amount Allocated",
        "Funding Cut", "Comments", "Email"
    ]].copy()

    flagged = flagged.sort_values(["Funding Cut", "Amount Requested"], ascending=[False, False])
    return flagged
```

## 9. Render the dashboard
Click **Refresh Dashboard** after changing filters.

```python
dashboard_output = widgets.Output()

def render_dashboard(_=None):
    filtered = apply_filters(df)

    with dashboard_output:
        clear_output(wait=True)

        if filtered.empty:
            display(HTML("<h3 style='color:#B91C1C;'>No rows match the current filters.</h3>"))
            return

        display(HTML("<h2 style='margin-bottom:6px;'>SAF Funding Dashboard</h2>"))
        display(HTML("<p style='margin-top:0; color:#4B5563;'>Frontend prototype built on the cleaned appropriations dataset.</p>"))
        display(HTML(make_kpi_html(filtered)))

        funnel_fig = make_funnel_chart(filtered)
        round_fig = make_round_chart(filtered)
        category_fig = make_category_chart(filtered)
        top_cios_fig = make_top_cios_chart(filtered)

        if funnel_fig is not None:
            funnel_fig.show()

        if round_fig is not None:
            round_fig.show()

        if category_fig is not None:
            category_fig.show()

        if top_cios_fig is not None:
            top_cios_fig.show()

        display(HTML("<h3>Audit Summary</h3>"))
        display(make_audit_summary(filtered))

        flagged = make_flagged_table(filtered)
        display(HTML("<h3>Flagged Rows</h3>"))
        if flagged.empty:
            display(HTML("<p>No audit flags in the current filtered view.</p>"))
        else:
            display(flagged.head(25))

        display(HTML("<h3>Filtered Data Preview</h3>"))
        display(filtered.head(25))

def reset_filters(_=None):
    category_widget.value = "All"
    round_widget.value = "All"
    semester_widget.value = "All"
    year_widget.value = "All"
    search_widget.value = ""
    top_n_widget.value = 10
    render_dashboard()

refresh_button.on_click(render_dashboard)
reset_button.on_click(reset_filters)

display(dashboard_output)
render_dashboard()
```
